```python
df['sources_list'] = df['source'].str.split('; ')

df.head(2)
```

```python
import numpy as np
from collections import Counter

flat_list = [item for sublist in df['sources_list'] for item in sublist]
source_counts = Counter(flat_list)
# sort source_counts
source_counts = {k: v for k, v in sorted(source_counts.items(), key=lambda item: item[1], reverse=True)}

source_counts
```

```python
source_df = pd.DataFrame.from_dict(source_counts, orient='index', columns=['Count'])
source_df = source_df.sort_values(by='Count', ascending=False).reset_index()
source_df.columns = ['Source', 'Count']
source_df['Rank'] = source_df['Count'].rank(ascending=False, method='min')
source_df['Percentage'] = source_df['Count'] / source_df['Count'].sum()
source_df
```
